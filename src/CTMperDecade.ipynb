```python
def clean_text_gtext(text,tokenizer,stopwords):
    tokens = tokenizer(text)
    tokens = [t for t in tokens if len(t) > 2]  # Remove short tokens
    tokens = [t for t in tokens if not t in stopwords]
    
    return " ".join(tokens)
```

```python
stop_words = stopwords.words("english")
sampled_df_train["clean_t"] =sampled_df_train["text"].map(lambda x: clean_text_gtext(x, word_tokenize, stop_words))
sampled_df_test["clean_t"] =sampled_df_test["text"].map(lambda x: clean_text_gtext(x, word_tokenize, stop_words))
```

```python
from contextualized_topic_models.models.ctm import ZeroShotTM
from contextualized_topic_models.utils.data_preparation import TopicModelDataPreparation
from contextualized_topic_models.utils.preprocessing import WhiteSpacePreprocessingStopwords
```

```python
def create_chunks(row):
    arr = row['clean_t'].split()
    n = 128
    chunks = [" ".join(arr[i:i+n]) for i in range(0, len(arr), 1152)]
    return pd.DataFrame({'chunk': chunks ,'document_id':row['document_id'],'decade':row['decade']})
```

```python
a = sampled_df_train.apply(lambda x: create_chunks(x),axis=1)
```

```python
chunks_df = pd.DataFrame()
for i,row in a.items():
    chunks_df = pd.concat([chunks_df,row])
```

```python
tp_models = []
models = []
training_sets = []
```

```python
stopwords_set = set(list(stop_words))
```

Pipeline:
1. Get the chunks
2. Preprocess them
3. Train the topic model data preparation embeddings 
4. Train the actual ZeroShotTM model, with 10 components
5. Save the preprocessing model, save the dataset, save the ZeroShotTM model

This can take a lot, it is training 8 models.

```python
stopwords_list = requests.get("https://gist.githubusercontent.com/rg089/35e00abf8941d72d419224cfd5b5925d/raw/12d899b70156fd0041fa9778d657330b024b959c/stopwords.txt").content
stopwords = set(stopwords_list.decode().splitlines())
stopwords.remove('thou')
```

```python
def pipeline(n_clusters = 10):
    tp_models = []
    models = []
    training_sets = []
    for i in range(1720,1800,10):
        documents = chunks_df[chunks_df['decade'] == i]["chunk"].tolist()
        sp = WhiteSpacePreprocessingStopwords(documents, stopwords_list=stopwords_set)
        sp.stopwords.update(stopwords)
        ecco_stop_list = ['rs', 'un', 'lie', 'ce', 'anti', 'lat', 'li', 'ir', 'et', 'pre', 'iv', 'tt', 'tl', 'de', 'tbe', 'ann', 'thle', 'viii', 'id', 'xii', 'lo', 'ibid', 'toe', 'xi', 'con', 'ti', 'er', 'vi', 'pro', 'vii', 'nd', 'wh', 'iz', 'zo', 'lr', 'non', 'ne', 'ii', 'le', 'ri', 'fr', 'en', 'al', 'tit', 'te', 'iii', 'cap', 'ij', 'til', 'hi', 'lon', 'fe', 'ib', 'tie', 'oo', 'la', 'di', 'pr', 'ie', 'ith', 'sect', 'ut', 'com', 'ol', 'ot', 'lib', 'ix', 'ex', 'par', 'ac', 'anp', 'ci', 'th', 'si', 'tr', 'quod', 'les', 'eft', 'que', 'bp', 'ed', 'ant', 'ano', 'ca', 'vide', 'es', 'du', 'lt', 'ch', 'vol', 'lit', 'ni', 'ditto', 'co', 'il', 'ro', 'cl', 'tile', 'ic', 'dc', 'cc', 'io', 'ai', 'per', 'ad', 'cr', 'chap', 'anno', 'tle', 'oi', 'ar', 'ill', 'inl', 'ly', 'thie', 'ha', 'geo fol', 'pa', 'andl', 'tha', 'thi', 'ta', 'mi', 'die', 'tlhe', 'stat', 'ihe', 'anld', 'qui', 'vel', 'ter', 'ver', 'ji', 'ani', 'thc', 'ili', 'regis', 'ld', 'anl', 'tlie', 'des', 'thec', 'andi', 'ire', 'andl', 'thi', 'fol', 'ta', 'pa', 'mi', 'ing', 'anld', 'andi', 'ihe', 'tha', 'tlhe', 'ion', 'tlie', 'geo', 'vel', 'ani', 'ld', 'car', 'ter', 'cum', 'ul', 'qui', 'ji', 'ich', 'ind', 'ate', 'anid', 'ere', 'thc','tte','lis']
        sp.stopwords.update(ecco_stop_list)
        preprocessed_documents, unpreprocessed_corpus, vocab, retained_indices = sp.preprocess()
        tp = TopicModelDataPreparation("paraphrase-multilingual-mpnet-base-v2")
        training_dataset = tp.fit(text_for_contextual=unpreprocessed_corpus, text_for_bow=preprocessed_documents)
        ctm = ZeroShotTM(bow_size=len(tp.vocab), contextual_size=768, n_components=n_clusters, num_epochs=20)
        ctm.fit(training_dataset)
        tp_models.append(tp)
        models.append(ctm)
        training_sets.append(training_dataset)
    return tp_models,models,training_sets
```

```python
tp_models,models,training_sets = pipeline(n_clusters = 10)
```

```python
test_chunks = sampled_df_test.apply(lambda x: create_chunks(x),axis=1)
chunks_df_test = pd.DataFrame()
for i,row in test_chunks.items():
    chunks_df_test = pd.concat([chunks_df_test,row])

test_sets = []
count = 0
for i in range(1720,1800,10):
    test_sets.append(tp_models[count].transform(chunks_df_test[chunks_df_test['decade'] == i]["chunk"].tolist()) )
    count += 1
```
