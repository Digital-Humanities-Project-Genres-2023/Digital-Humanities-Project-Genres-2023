```python
import panel as pn
import pandas as pd

pn.extension("tabulator")
```

```python
import sys
sys.path.append("../")
from src import puhti_files
```

```python
df = puhti_files.genre_data_to_pandas(data="dev")
```

```python
df.head()
```

```python
table = pn.widgets.Tabulator(
    df,
    disabled=True,
    show_index=False,
    hidden_columns=["document_id"],
    selection=[0],
    selectable=1,
    #layout='fit_data',
    #width=500,
    widths={
        "work_id": "80%",
        "main_category": "10%",
        "sub_category": "10%"
    },
    #sizing_mode='stretch_width'
)
text = pn.pane.Str(puhti_files.read_text_file(df["document_id"][0]))
```

```python
def click(event):
    text[0].object = puhti_files.read_text_file(df["document_id"][event.row])
```

```python
table.on_click(click)
```

```python
gspec = pn.GridSpec(sizing_mode='stretch_both', max_height=800)
gspec[0,0] = table
gspec[0, 1:3] = text
```
