```python
df = puhti_files.genre_data_to_pandas(data="test")
len(df)
```

```python
ecco_metadata = puhti_files.ecco_metadata_to_pandas()
ecco_metadata = ecco_metadata.set_index("document_id_octavo")
```

```python
labels = puhti_files.labels_to_dict(invert=True)
```

```python
import time
import os
import pandas as pd

counter = 0
start_time = time.time()
zip_file = "../books_by_categories.zip"

with zipfile.ZipFile(zip_file, mode="a", compression=zipfile.ZIP_DEFLATED) as my_zip:
    for idx, row in df.iterrows():
        doc_id = row['document_id']
        file_name = f"{doc_id}.txt"
        file_path = '/scratch/project_2007227/genre_data/ecco_source/' + file_name
        main_cat = row['main_category']
        sub_cat = row['sub_category']
        zip_path = f"{main_cat}_{labels['main2id'][main_cat]}/{sub_cat}_{labels['sub2id'][sub_cat]}/{doc_id}"
        
        #Book Text
        my_zip.write(
            file_path,
            arcname=f"{zip_path}/{file_name}"
        )
        
        #File Info
        my_zip.writestr(
            f"{zip_path}/book_info.json",
            row.to_json(indent=2)
        )
        
        #Ecco Metadata
        ecco_row = ecco_metadata.loc[doc_id]
        if isinstance(ecco_row, pd.DataFrame) and ecco_row.shape[0] > 1:
            for i in range(0, len(ecco_row)):
                my_zip.writestr(
                    f"{zip_path}/ecco_metadata_{i+1}.json",
                    ecco_row.iloc[i].to_json(indent=2)
                )
        else:
            my_zip.writestr(
                f"{zip_path}/ecco_metadata_1.json",
                ecco_row.to_json(indent=2)
            )
        
        counter += 1

        # Print information for every 100 rows
        if counter % 1000 == 0:
            rows_done = counter
            rows_remaining = len(df) - counter
            time_elapsed = time.time() - start_time
            time_per_row = time_elapsed / rows_done
            estimated_time_remaining = time_per_row * rows_remaining
            
            zip_file_size_mb = os.path.getsize(zip_file) / (1024 * 1024)
            size_per_row = zip_file_size_mb / rows_done
            estimated_total_size = size_per_row * len(df)

            print(f'Rows done: {rows_done}')
            print(f'Rows remaining: {rows_remaining}')
            print(f'Time elapsed: {time_elapsed:.2f} seconds')
            print(f'Estimated time remaining: {estimated_time_remaining:.2f} seconds')
            print(f"The size of '{zip_file}' is {zip_file_size_mb:.2f} MB.")
            print(f"Size per row: {size_per_row:.2f} MB")
            print(f"Estimated total size: {estimated_total_size:.2f} MB")
            print('-' * 10)
```
